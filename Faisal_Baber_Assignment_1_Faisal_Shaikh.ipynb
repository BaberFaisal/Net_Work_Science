### Task 2. Nodes labeling (0 points)

Nodes in many real-world graphs represents some objects, say people in social networks or atoms in molecular networks. Usually, names of such nodes is in the string format, but in many cases it can be useful to make names and indices of nodes equal. Using such labeling, we can refer to a particular node by its index.

Write a function `convert_node_names_to_int` that takes a graph with string names and returns a new graph with integer names so that first node in the initial graph is called 0 in the new graph, second is 1 and so on. Also, save initial names in the node attribute `name`.

````python
def convert_node_names_to_int(G):
    # YOUR CODE HERE
    #raise NotImplementedError()


    mapping = {}
    for ii, name in enumerate(G.nodes):
      mapping[name] = ii

    newG = nx.Graph()
    for name in G.nodes:
      newG.add_node(mapping[name])
      newG.nodes[mapping[name]]['name'] = name

    for edge in G.edges:
      u, v = edge
      newG.add_edge(mapping[u], mapping[v])

    return newG
````

Write a function `convert_to_digraph` that takes an undirected graph and returns a new directed graph where the direction is defined by the rule: the node $i$ follows to $j$ if the length of `name` of $i$ is at least the length of `name` of $j$. Keep the ordering of nodes and the node attribute `name`.

````python
def convert_to_digraph(G):
    # YOUR CODE HERE
    #raise NotImplementedError()

    newG = nx.DiGraph()
    newG.add_nodes_from(G.nodes)
    nx.set_node_attributes(newG, values=nx.get_node_attributes(G, "name"), name='name')

    for edge in G.edges:
      i, j = edge
      if len(G.nodes[i]['name']) > len(G.nodes[j]['name']):
        newG.add_edge(i, j)
      elif len(G.nodes[i]['name']) < len(G.nodes[j]['name']):
        newG.add_edge(j, i)
      else:
        newG.add_edge(i, j)
        newG.add_edge(j, i)

    return newG
````

Your task is to draw the matrix graph using the kamada kawai layout with nonlinear modification to obtain a denser layout away from the NEO node. Let us define a proximity coefficient using Gaussian kernel

$$\text{coef}(i) = \exp\left(-\frac{\text{distance}^2}{2\sigma^2}\right)$$

where a _distance_ is the Euclidean distance from the node $i$ to the NEO node, $\sigma$ is the constant. Then the nodes should be moved towards the NEO node such that

$$\text{distance} = \text{coef}(i) \times \text{distance}$$

The simplest way to do it is to center the layout so that the NEO node is in x=0 y=0 and then to multiply all node positions by the proximity coefficient.

Write a function `dense_neo_layout` that takes a layout, sigma and returns a new (dense) layout.

_Hint: use `np.linalg.norm` to calculate Euclidean norm._

````python
def dense_neo_layout(layout, sigma):
    # Centering the layout
    neo_pos = layout['NEO']
    centered_layout = {node: np.array(pos) - np.array(neo_pos) for node, pos in layout.items()}

    # Calculating  coefficient
    dense_layout = {}
    for node, pos in centered_layout.items():
        distance = np.linalg.norm(pos)
        if distance == 0:
            coef = 1
        else:
            coef = np.exp(-distance**2 / (2 * sigma**2))
        dense_layout[node] = pos * coef

    return dense_layout

    raise NotImplementedError()
````

Write a function `drawing_nodes_params` that takes a graph and returns a dictionary with parameters for `nx.draw_networkx` of the form `{'G': graph, 'node_color': color, ...}`. Parameters should contain a graph, kamada kawai layout, node colors from `viz.color` attribute. The colors should be converted from RGB range 0-255 to RGB range 0-1. The node size is a node degree `G.degree[node]` multiplied by 100.

````python
def prepare_node_visualization(graph):
    """
    Prepares parameters for drawing a graph with custom node colors and sizes.
    """
    # Compute the Kamada-Kawai layout for the graph
    node_positions = nx.kamada_kawai_layout(graph)

    # Prepare node colors
    colors = []
    for node in graph.nodes:
        # Retrieve color information from node attributes or default to white
        color_data = graph.nodes[node].get('viz', {}).get('color', {'r': 255, 'g': 255, 'b': 255})
        # Normalize the color values to [0, 1]
        normalized_color = np.array([color_data['r'] / 255, color_data['g'] / 255, color_data['b'] / 255])
        colors.append(normalized_color)

    # Calculate node sizes based on their degrees
    sizes = [graph.degree[n] * 100 for n in graph.nodes]

    # Return a dictionary containing all parameters for visualization
    return {
        'graph': graph,
        'positions': node_positions,
        'node_colors': colors,
        'node_sizes': sizes
    }
````

Write a function `drawing_nodes_polyparams` that returns the same dictionary as the `drawing_nodes_params` function, but the node size is a sum of node degree in a power 3 and 10.

$$\text{size} = \text{degree}^3 + 10$$

````python
def generate_visualization_parameters(graph):
    """
    Generates parameters for drawing a graph with customized node colors and sizes.
    """
    # Compute node positions using Kamada-Kawai layout
    node_positions = nx.kamada_kawai_layout(graph)

    # Process node colors
    color_list = []
    for node in graph.nodes:
        # Retrieve color from node attributes or default to white
        color_info = graph.nodes[node].get('viz', {}).get('color', {'r': 255, 'g': 255, 'b': 255})
        # Normalize RGB values to the [0, 1] range
        normalized_rgb = [color_info['r'] / 255, color_info['g'] / 255, color_info['b'] / 255]
        color_list.append(normalized_rgb)

    # Compute node sizes based on the cube of the degree (non-linear scaling)
    size_list = [(graph.degree[node] ** 3) + 10 for node in graph.nodes]

    # Return visualization parameters as a dictionary
    return {
        'graph': graph,
        'positions': node_positions,
        'node_colors': color_list,
        'node_sizes': size_list
    }
````

Write a function `labels_list_parameters` that takes a graph, layout and returns a list of parameters for repeated drawing labels using `nx.draw_networkx_labels`. Each element of the list is a dictionary that can be passed to `nx.draw_networkx_labels`. The first dictionary is for the node degree 6, the second is for 7 and so on.

````python
# Used Chatgpt to take some help
def generate_label_parameters(graph, layout):
    """
    Generates parameters for labeling graph nodes based on their degree.

    Nodes with a degree >= 6 are grouped, and their label positions are adjusted.
    """
    # Filter nodes based on degree threshold
    high_degree_nodes = [node for node in graph.nodes if graph.degree[node] >= 6]

    # Group nodes by their degree
    degree_groups = {}
    for node in high_degree_nodes:
        degree = graph.degree[node]
        degree_groups.setdefault(degree, []).append(node)

    # Create a list of parameters for each degree group
    label_params_list = []
    for degree, grouped_nodes in sorted(degree_groups.items()):
        adjusted_positions = {}

        # Adjust label positions based on degree
        for node in grouped_nodes:
            x, y = layout[node]
            y_adjusted = y - 0.025 * np.sqrt(degree)  # Fine-tune y-position
            adjusted_positions[node] = np.array([x, y_adjusted])

        # Create label mapping for the current group
        labels = {node: str(node) for node in grouped_nodes}

        # Append the parameters for this group
        label_params_list.append({
            'graph': graph,
            'positions': adjusted_positions,
            'labels': labels,
            'font_size': degree
        })

    return label_params_list
````

Write a function `get_all_params` that takes a graph and layout and returns a dictionary with keys:

* `labels_param_list` — list of dict, parameters for `nx.draw_networkx_edges`, for degrees 6, 7 and so on
* `area` — list of float vlues: a left, right, lower, higher borders of the visible area

````python
def get_all_params(graph, layout):
    # YOUR CODE HERE
    node_colors = []
    node_sizes = []
    for node in graph.nodes:
        # Extract node color from 'viz.color' attribute
        color = graph.nodes[node].get('viz', {}).get('color', {'r': 255, 'g': 255, 'b': 255})  # Default to white if no color
        normalized_color = np.array([color['r'] / 255, color['g'] / 255, color['b'] / 255])  # Convert to NumPy array for rounding
        node_colors.append(normalized_color)

        # Calculate node size as 5 * degree^2
        degree = graph.degree[node]
        node_sizes.append(5 * degree ** 2)

    nodes_params = {
        'G': graph,
        'pos': layout,
        'node_color': node_colors,
        'node_size': node_sizes
    }

    # Edge parameters
    weights = nx.get_edge_attributes(graph, 'weight')
    edge_widths = list(weights.values())

    edges_params = {
        'G': graph,
        'pos': layout,
        'width': edge_widths,
        'alpha': 0.2
    }

    # Label parameters
    labels_param_list = []
    nodes_with_degree_6_plus = [node for node in graph.nodes if graph.degree[node] >= 6]
    for node in nodes_with_degree_6_plus:
        degree = graph.degree[node]
        font_size = 3 * degree ** (2/3)

        # Adjust label position
        x, y = layout[node]
        y_adjusted = y - 0.025 * np.sqrt(degree)
        adjusted_positions = {node: np.array([x, y_adjusted])}

        labels_param_list.append({
            'G': graph,
            'pos': adjusted_positions,
            'labels': {node: str(node)},
            'font_size': font_size
        })

    # Visible area
    x_coords = [layout[node][0] for node in graph.nodes]
    y_coords = [layout[node][1] for node in graph.nodes]

    # Define borders based on specific nodes
    left = layout['MORPHEUS'][0]
    right = layout['TANK'][0]
    lower = layout['NEO'][1]
    higher = layout['AGENT SMITH'][1]

    area = [left, right, lower, higher]

    # Return all parameters
    return {
        'nodes_params': nodes_params,
        'edges_params': edges_params,
        'labels_param_list': labels_param_list,
        'area': area
    }
    raise NotImplementedError()
````
